```python
google_app_sans_NaN = google_app.dropna()
```

```python
google_app_sans_NaN.shape
```

```python
google_app_sans_NaN["Installs"]=google_app_sans_NaN["Installs"].apply(lambda x: x.replace(",", ""))
google_app_sans_NaN["Installs"]=google_app_sans_NaN["Installs"].apply(lambda x: x.replace("+", ""))
```

```python
google_app_sans_NaN["Installs"]=google_app_sans_NaN["Installs"].apply(lambda x: int(x))
```

```python
google_app_sans_NaN["Reviews"]=google_app_sans_NaN["Reviews"].apply(lambda x: int(x))
```

```python
google_app_sans_NaN["Price"]=google_app_sans_NaN["Price"].apply(lambda x: x.replace("$", ""))
google_app_sans_NaN["Price"]=google_app_sans_NaN["Price"].apply(lambda x: float(x))
```

```python
#1. Quels sont les noms, les catégories et les genres des applications les plus installées ?
max_install=google_app_sans_NaN["Installs"].max()
```

```python
max_install
```

```python
google_app_sans_NaN.loc[google_app_sans_NaN["Installs"] == max_install, ["App", "Category", "Genres"]]
```

```python
#2. Quel est le plus grand nombre de Reviews ? Quel est le nom et la note globale de l'application concernée ?
plus_grand_nombre_de_reviews=google_app_sans_NaN["Reviews"].max()
print(plus_grand_nombre_de_reviews)
```

```python
google_app_sans_NaN.loc[google_app_sans_NaN["Reviews"] == plus_grand_nombre_de_reviews, ["App", "Rating"]]
```
