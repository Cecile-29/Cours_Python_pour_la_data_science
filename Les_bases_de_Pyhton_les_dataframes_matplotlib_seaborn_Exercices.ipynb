```python
############# EXERCICE Visualisation des données ############

# Objectif : Explorer les additions encaissées par un restaurant

# Importer les librairies à utiliser
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le jeu de données "tips.csv"
jeu_de_donnees_tips = pd.read_csv("tips.csv")
jeu_de_donnees_tips
```

```python
# charger le jeu de données 
tips=sns.load_dataset("tips")
```

```python
# 1 Tracer un barplot avec la fonction sns.barplot représentant la variable total_bill et en séparant ces données par sexe
sns.barplot(x='sex', y='total_bill', data=tips)
# sns.barplot(x='sex', y='total_bill', hue='smoker', data=tips)
```

```python
# Tracer un barplot avec total_bill et séparation par sexe
sns.barplot(x='sex', y='total_bill', data=tips, palette={'Male': 'blue', 'Female': 'red'})

# Ajouter des étiquettes et un titre
plt.xlabel("Sexe")
plt.ylabel("Total Bill")
plt.title("Total Bill par sexe")

# Afficher le graphique
plt.draw()
```

```python
# 2 Utiliser la fonction sns.distplot sur la variable total_bill pour étudier sa distribution
sns.distplot(tips['total_bill'], kde=False)
```

```python
###########################         sns.distplot(tips['total_bill'], kde=False)                     ########################

''' La fonction sns.distplot() de la bibliothèque Seaborn est utilisée pour tracer un histogramme et une estimation de la densité de probabilité (courbe KDE) d’une variable continue. Voici une explication plus détaillée :

Histogramme :
L’histogramme est un graphique qui représente la distribution des valeurs d’une variable.
Il divise la plage de valeurs en intervalles (ou “bins”) et compte le nombre d’observations dans chaque intervalle.
Chaque barre verticale de l’histogramme représente la fréquence (ou le nombre d’observations) dans un intervalle donné.
Courbe KDE (Kernel Density Estimation) :
La courbe KDE est une estimation de la densité de probabilité de la variable.
Elle estime la distribution continue des données en utilisant une fonction noyau (ou “kernel”) pour chaque observation.
La courbe KDE estime la forme de la distribution sans discrétiser les données en intervalles comme le fait l’histogramme.
Utilisation de sns.distplot() :
Dans votre exemple, sns.distplot(tips['total_bill'], kde=False) :
tips['total_bill'] est la colonne “total_bill” du DataFrame “tips”.
kde=False désactive la courbe KDE (si vous souhaitez uniquement l’histogramme).
En résumé, sns.distplot() combine un histogramme et une estimation de la densité de probabilité pour visualiser la distribution d’une variable continue. Si vous avez d’autres questions, n’hésitez pas à me demander ! 😊

: Documentation Seaborn - Distribution Plots '''
```

```python
# 3 Utiliser la fonction jointplot de Seaborn pour tracer la variable total_bill (X) en fonction de la variable tip (Y) 
sns.jointplot(data=tips, x="total_bill", y="tip")
plt.draw()
```

```python
'''    
Jointplot :  La fonction sns.jointplot() de la bibliothèque Seaborn permet de créer des graphiques de dispersion (ou scatter plots) pour explorer la relation entre deux variables continues. 

Graphique de dispersion (scatter plot) :
Un graphique de dispersion affiche les points correspondant aux valeurs de deux variables sur un plan cartésien.
L’axe des x représente une variable  “total_bill”, et l’axe des y représente une autre variable “tip”.
Chaque point sur le graphique représente une paire de valeurs (x, y) pour chaque observation dans vos données.
Utilisation de  sns.jointplot(data=tips, x="total_bill", y="tip") :
data=tips est le DataFrame contenant vos données.
x="total_bill" et y="tip" spécifient les colonnes à utiliser pour l’axe des x et l’axe des y, respectivement.
Cela  permet d’explorer la relation entre deux variables continues en affichant un graphique de dispersion.'''
```

```python
# 4 Utiliser la fonction pairplot de Seaborn sur l’ensemble du dataframe « tips » pour étudier les relations entre l’ensemble des variables.
sns.pairplot(tips)
```

```python
''' La fonction sns.pairplot() sert à créer des graphiques de dispersion pour explorer les relations entre toutes les paires de variables continues dans un jeu de données. 

Chaque variable numérique est comparée à toutes les autres variables ###numériques###, et les graphiques de dispersion sont affichés sur la diagonale pour montrer la distribution de chaque variable individuelle.'''
```

```python
# 5 Tracer le boxplot de la variable total_bill en créant un boxplot par jour (variable day) et en utilisant la variable sex dans l’option hue.

sns.boxplot(x="day", y="total_bill", hue="sex", data=tips)
```

```python
''' La fonction sns.boxplot() ser à créer des boxplots (ou diagrammes en boîte) pour explorer la distribution des valeurs d’une variable continue, en tenant compte de différentes catégories. 
Boxplot (diagramme en boîte) :
Un boxplot est un graphique qui résume la distribution des valeurs d’une variable.
Il affiche les quartiles (Q1, médiane et Q3) ainsi que les valeurs aberrantes potentielles.
La boîte représente l’intervalle interquartile (IQR), la ligne à l’intérieur de la boîte est la médiane, et les moustaches (whiskers) représentent l’étendue des données.
Les points individuels à l’extérieur des moustaches sont les valeurs aberrantes.
Utilisation de sns.boxplot() :
Dans l'exemple, sns.boxplot(x="day", y="total_bill", hue="sex", data=tips) :
x="day" spécifie la variable pour l’axe des x (jours de la semaine).
y="total_bill" spécifie la variable pour l’axe des y (montant total de la facture).
hue="sex" sépare les données par sexe (féminin ou masculin).  '''
```
