```python
############# EXERCICE Visualisation des donn√©es ############

# Objectif : Explorer les additions encaiss√©es par un restaurant

# Importer les librairies √† utiliser
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le jeu de donn√©es "tips.csv"
jeu_de_donnees_tips = pd.read_csv("tips.csv")
jeu_de_donnees_tips
```

```python
# charger le jeu de donn√©es 
tips=sns.load_dataset("tips")
```

```python
# 1 Tracer un barplot avec la fonction sns.barplot repr√©sentant la variable total_bill et en s√©parant ces donn√©es par sexe
sns.barplot(x='sex', y='total_bill', data=tips)
# sns.barplot(x='sex', y='total_bill', hue='smoker', data=tips)
```

```python
# Tracer un barplot avec total_bill et s√©paration par sexe
sns.barplot(x='sex', y='total_bill', data=tips, palette={'Male': 'blue', 'Female': 'red'})

# Ajouter des √©tiquettes et un titre
plt.xlabel("Sexe")
plt.ylabel("Total Bill")
plt.title("Total Bill par sexe")

# Afficher le graphique
plt.draw()
```

```python
# 2 Utiliser la fonction sns.distplot sur la variable total_bill pour √©tudier sa distribution
sns.distplot(tips['total_bill'], kde=False)
```

```python
###########################         sns.distplot(tips['total_bill'], kde=False)                     ########################

''' La fonction sns.distplot() de la biblioth√®que Seaborn est utilis√©e pour tracer un histogramme et une estimation de la densit√© de probabilit√© (courbe KDE) d‚Äôune variable continue. Voici une explication plus d√©taill√©e :

Histogramme :
L‚Äôhistogramme est un graphique qui repr√©sente la distribution des valeurs d‚Äôune variable.
Il divise la plage de valeurs en intervalles (ou ‚Äúbins‚Äù) et compte le nombre d‚Äôobservations dans chaque intervalle.
Chaque barre verticale de l‚Äôhistogramme repr√©sente la fr√©quence (ou le nombre d‚Äôobservations) dans un intervalle donn√©.
Courbe KDE (Kernel Density Estimation) :
La courbe KDE est une estimation de la densit√© de probabilit√© de la variable.
Elle estime la distribution continue des donn√©es en utilisant une fonction noyau (ou ‚Äúkernel‚Äù) pour chaque observation.
La courbe KDE estime la forme de la distribution sans discr√©tiser les donn√©es en intervalles comme le fait l‚Äôhistogramme.
Utilisation de sns.distplot() :
Dans votre exemple, sns.distplot(tips['total_bill'], kde=False) :
tips['total_bill'] est la colonne ‚Äútotal_bill‚Äù du DataFrame ‚Äútips‚Äù.
kde=False d√©sactive la courbe KDE (si vous souhaitez uniquement l‚Äôhistogramme).
En r√©sum√©, sns.distplot() combine un histogramme et une estimation de la densit√© de probabilit√© pour visualiser la distribution d‚Äôune variable continue. Si vous avez d‚Äôautres questions, n‚Äôh√©sitez pas √† me demander ! üòä

: Documentation Seaborn - Distribution Plots '''
```

```python
# 3 Utiliser la fonction jointplot de Seaborn pour tracer la variable total_bill (X) en fonction de la variable tip (Y) 
sns.jointplot(data=tips, x="total_bill", y="tip")
plt.draw()
```

```python
'''    
Jointplot :  La fonction sns.jointplot() de la biblioth√®que Seaborn permet de cr√©er des graphiques de dispersion (ou scatter plots) pour explorer la relation entre deux variables continues. 

Graphique de dispersion (scatter plot) :
Un graphique de dispersion affiche les points correspondant aux valeurs de deux variables sur un plan cart√©sien.
L‚Äôaxe des x repr√©sente une variable  ‚Äútotal_bill‚Äù, et l‚Äôaxe des y repr√©sente une autre variable ‚Äútip‚Äù.
Chaque point sur le graphique repr√©sente une paire de valeurs (x, y) pour chaque observation dans vos donn√©es.
Utilisation de  sns.jointplot(data=tips, x="total_bill", y="tip") :
data=tips est le DataFrame contenant vos donn√©es.
x="total_bill" et y="tip" sp√©cifient les colonnes √† utiliser pour l‚Äôaxe des x et l‚Äôaxe des y, respectivement.
Cela  permet d‚Äôexplorer la relation entre deux variables continues en affichant un graphique de dispersion.'''
```

```python
# 4 Utiliser la fonction pairplot de Seaborn sur l‚Äôensemble du dataframe ¬´¬†tips¬†¬ª pour √©tudier les relations entre l‚Äôensemble des variables.
sns.pairplot(tips)
```

```python
''' La fonction sns.pairplot() sert √† cr√©er des graphiques de dispersion pour explorer les relations entre toutes les paires de variables continues dans un jeu de donn√©es. 

Chaque variable num√©rique est compar√©e √† toutes les autres variables ###num√©riques###, et les graphiques de dispersion sont affich√©s sur la diagonale pour montrer la distribution de chaque variable individuelle.'''
```

```python
# 5 Tracer le boxplot de la variable total_bill en cr√©ant un boxplot par jour (variable day) et en utilisant la variable sex dans l‚Äôoption hue.

sns.boxplot(x="day", y="total_bill", hue="sex", data=tips)
```

```python
''' La fonction sns.boxplot() ser √† cr√©er des boxplots (ou diagrammes en bo√Æte) pour explorer la distribution des valeurs d‚Äôune variable continue, en tenant compte de diff√©rentes cat√©gories. 
Boxplot (diagramme en bo√Æte) :
Un boxplot est un graphique qui r√©sume la distribution des valeurs d‚Äôune variable.
Il affiche les quartiles (Q1, m√©diane et Q3) ainsi que les valeurs aberrantes potentielles.
La bo√Æte repr√©sente l‚Äôintervalle interquartile (IQR), la ligne √† l‚Äôint√©rieur de la bo√Æte est la m√©diane, et les moustaches (whiskers) repr√©sentent l‚Äô√©tendue des donn√©es.
Les points individuels √† l‚Äôext√©rieur des moustaches sont les valeurs aberrantes.
Utilisation de sns.boxplot() :
Dans l'exemple, sns.boxplot(x="day", y="total_bill", hue="sex", data=tips) :
x="day" sp√©cifie la variable pour l‚Äôaxe des x (jours de la semaine).
y="total_bill" sp√©cifie la variable pour l‚Äôaxe des y (montant total de la facture).
hue="sex" s√©pare les donn√©es par sexe (f√©minin ou masculin).  '''
```
