```python
# exemple de graphique avec les données de sommes de ventes par catégorie d'âge dans black_friday
# création d'un dictionnaire qui sera une structure de données adaptée
Age_groupe = black_friday.groupby("Age")
Age_groupe
# en sortie --> un objet
```

```python
# méthode count()
Age_groupe.count()
```

```python
# initialisation d'un dictionnaire vide avec la méthode dict()
achats_clients=dict()

# boucle pour création d'un dictionnaire avec les données de ventes par catégorie d'âge dans black_friday
for age,group in Age_groupe:
    achats_clients[age]=sum(group["Purchase"])
```

```python
achats_clients
```

```python
# explications : la boucle for parcourt chaque élément de Age_groupe, où age est la clé (groupe d’âge) et group est la valeur (DataFrame contenant les données de ventes).
# À chaque itération, la somme des achats (sum(group["Purchase"])) est ajoutée au dictionnaire achats_clients avec la clé correspondante (age).
# l'itération se fait elle sur un sous ensemble  du dataframe à travers les groupes d’âge et leurs données de ventes, on ajoute la somme des achats pour chaque groupe au dictionnaire achats_clients
for age,group in Age_groupe:
    print(age)
    print(group)
```

```python
# visualisation des données de ventes par catégorie d'âge dans black_friday avec plt.plot()
plt.plot(achats_clients.keys(), achats_clients.values())
# pour la visualisation lsur le graphiques la clé keys() est valeur de x (abscisse) et la valeur .values() de y (ordonée)
plt.draw()
```

```python
# ajouter de la couleur au graphique avec   scatter()
# avant d'afficher la fenêtre graphique on ajoute un paramètre rcParams pour changer la valeur de figure.figsize et taille à 16,9 
plt.rcParams["figure.figsize"]=[16,9]
plt.scatter(achats_clients.keys(), achats_clients.values())
plt.draw()
```

```python
# changer le dictionnaire achats_clients en un dataframe en liste pour la ft scatter qui attend une liste
# c=list sert à colorer les points du graphique selon la somme dépensée par la catégorie d'âge
plt.rcParams["figure.figsize"]=[16,9]
plt.scatter(achats_clients.keys(), achats_clients.values(), c=list(achats_clients.values()))
plt.xlabel("Catégorie d'âge")
plt.ylabel("total des ventes")
plt.title("Somme totale des ventes effectuées par catégories d'âge", y=1.05)
plt.draw
```

```python
# mettre la couleur du bleu au rouge selon la somme des achats pour les points du graphique
# aide matplotlib colorReference sur la doc seismic fonctionne avec cmap()
plt.rcParams["figure.figsize"]=[16,9]
plt.scatter(achats_clients.keys(), achats_clients.values(), c=list(achats_clients.values()), cmap="seismic")
plt.xlabel("Catégorie d'âge")
plt.ylabel("total des ventes")
plt.title("Somme totale des ventes effectuées par catégories d'âge", y=1.05)
plt.draw
```

```python
# kind="swarm" pour afficher les points distincts   
sns.catplot(x="Age", y="Purchase", hue="Age", data=subset, kind="swarm")
plt.draw
plt.savefig("mon_plus_beau_graphique.png", bbox_inches="tight")
```
