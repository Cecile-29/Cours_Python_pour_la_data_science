```python
# 2 Quel est le plus grand nombre de Reviews ?

#  d’abord convertir la colonne “Reviews” en un type numérique, cactuellement de type object, généralement une chaîne de caractères. 
# Une fois convertie, utiliser la méthode .max() pour obtenir la valeur maximale. 

# Convertir la colonne 'Reviews' en entier
google_app_sans_NaN['Reviews'] = google_app_sans_NaN['Reviews'].astype(int)

# Trouver le maximum de la colonne 'Reviews'
max_reviews = google_app_sans_NaN['Reviews'].max()

# Afficher le résultat
print("Le plus grand nombre de reviews est :", max_reviews)
```

```python
# 3 Combien y a-t-il d’applications ouvertes aux personnes de tous âge ? (Everyone)

# Pour compter le nombre d’applications dans le DataFrame google_app_sans_NaN qui sont ouvertes aux personnes de tous âges, c’est-à-dire avec la mention “Everyone” dans la colonne “Content Rating”,
#  vous pouvez utiliser la méthode value_counts()


nb_apps_everyone = google_app_sans_NaN["Content Rating"].value_counts()["Everyone"]
print("Nombre d'applications pour 'Everyone' :", nb_apps_everyone)
```

```python
# 3 Combien y a-t-il d’applications ouvertes aux personnes de tous âge ? (Everyone)
 # ou ou combiner les méthodes loc[] et count()

nb_apps_everyone = google_app_sans_NaN.loc[google_app_sans_NaN["Content Rating"] == "Everyone", "Content Rating"].count()
print("Nombre d'applications pour 'Everyone' :", nb_apps_everyone)
```

```python
# Dans quelle catégorie y a-t-il le plus d’applications ?
# Pour déterminer dans quelle catégorie il y a le plus d’applications utiliser la méthode value_counts() sur la colonne “Category”. 
#  Cela comptera le nombre d’occurrences de chaque catégorie unique et triera les résultats par ordre décroissant.

# Compter le nombre d'applications par catégorie
categorie_counts = google_app_sans_NaN['Category'].value_counts()

# Trouver la catégorie avec le plus grand nombre d'applications
top_categorie = categorie_counts.idxmax()

print(f"La catégorie avec le plus d'applications est : {top_categorie}")
```

```python
# Quelle est l’application vendue la plus chère ?
# Pour déterminer l’application la plus chère, trier le DataFrame google_app_sans_NaN par la colonne “Price” en ordre décroissant 
# et sélectionner la première ligne index[0]

# Trier le DataFrame par la colonne "Price" en ordre décroissant
google_app_sans_NaN_sorted = google_app_sans_NaN.sort_values(by="Price", ascending=False)

# Sélectionner la première ligne (l'application la plus chère)
app_la_plus_chere = google_app_sans_NaN_sorted.iloc[0]

# Afficher le nom de l'application et son prix
print(f"L'application la plus chère est : {app_la_plus_chere['App']} avec un prix de {app_la_plus_chere['Price']}")
```
